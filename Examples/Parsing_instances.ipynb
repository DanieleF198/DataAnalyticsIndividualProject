This program just parse the istances obtained by the link https://data.mendeley.com/datasets/wy5bjtt47v/1 in a readible file gml, so that they are usable with networkx python package. To understand the format of starting files read readme.txt file in "Problems" folder. I have converted only graph with 50 nodes (markets) and 50 product to avoid to much large problem (and so too much time of execution for their resolution).

```python
import os
import networkx as nx
import numpy as np

path = './Problems/TPP_FSO/'
file_counter = 1
for filename in os.listdir(path):
    f = os.path.join(path, filename)
    if os.path.isfile(f):
        # READ FILE

        fG = open(f)
        data = fG.read()
        fG.close()
        lines = data.split('\n')
        V = 0   # number of nodes
        K = 0   # number of products
        D = 0   # time limit
        for i, line in enumerate(lines):
            if i == 0:
                continue   # not interested in title of the problem
            if i == 1:
                V = int(line[5:len(line)])
            if i == 2:
                K = int(line[5:len(line)])
            if i == 3:
                D = int(line[3:len(line)])
            if i == 4:
                break
        if V != 50 or K != 50:
            continue
        node_coord = np.zeros((V, 2), dtype="int32")   # coordinates (x,y) of nodes
        demand_product = np.zeros((K, 1), dtype="int32")   # demand for product
        offer_product = np.zeros((V, K, 2), dtype="int32")   # product offered by markets
        service_time = np.zeros((V, 2), dtype="int32")   # service time both for standard and fast service
        fast_cost = np.zeros((V, 1), dtype="int32")   # cost of fast service
        count_coord = 0
        count_demand = 0
        count_offer = 1   # depot has value 0, so start directly from second node
        count_time = 1
        count_fast = 1
        for i, line in enumerate(lines):
            if i < 5:
                continue
            if (i >= 5) and (i < 5 + V):
                values = [x for x in line.split(' ')[1:]]
                node_coord[count_coord, :] = values
                count_coord += 1
            if (i > 6 + V) and (i <= 6 + 2 * V):
                values = [x for x in line.split(' ')[1:]]
                demand_product[count_demand, :] = values
                count_demand += 1
            if (i > 8 + 2 * V) and (i < 8 + 3 * V):
                number_of_product = int(line.split(' ')[1])
                values = [x for x in line.split(' ')][3:]
                del values[-1]
                for k, element in enumerate(values):
                    if k%3 == 0:
                        offer_product[count_offer, int(element)-1,:] = [values[k+1], values[k+2]]
                count_offer += 1
            if (i > 9 + 3 * V) and (i < 9 + 4 * V):
                values = [x for x in line.split(' ')[1:]]
                service_time[count_time, :] = values
                count_time += 1
            if (i > 10 + 4 * V) and (i < 10 + 5 * V):
                values = [x for x in line.split(' ')[1:]]
                fast_cost[count_fast, :] = values
                count_fast += 1
        # CREATE GRAPH
        
        G = nx.Graph()
        for i in range(0, V):
            G.add_node(i, x=str(node_coord[i, 0]), y=str(node_coord[i, 1]), standard_time=str(service_time[i, 0]), fast_time=str(service_time[i, 1]), fast_cost=str(fast_cost[i, 0]), market_cost=str(offer_product[i,:,0]).replace('\n', ''), market_availability=str(offer_product[i,:,1]).replace('\n', ''))
        for i in range(0, V):
            for j in range(0, V):
                if i != j:
                    G.add_edge(i, j, distance=str(int(np.sqrt(pow(int(G.nodes[i]['x'])-int(G.nodes[j]['x']), 2)+ pow(int(G.nodes[i]['y'])-int(G.nodes[j]['y']), 2)))))
                    
        nx.write_gml(G, "./Graphs/TPP_FSO_CapEuclideo_" + str(V) + "_" + str(K) + "_" + str(file_counter) +".gml")
        np.savetxt("./Demand/TPP_FSO_CapEuclideo_" + str(V) + "_" + str(K) + "_" + str(file_counter) +".csv", demand_product, delimiter=',') # I save demand of each product in a different folder just for organize better the data
        with open("./Time_limits/TPP_FSO_CapEuclideo_" + str(V) + "_" + str(K) + "_" + str(file_counter) +".txt", 'w+') as f:
            f.write(str(D))
        file_counter += 1
```
